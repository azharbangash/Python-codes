```python
import numpy as np
import matplotlib.pyplot as plt

amp = 1
freq = 1
duration = 2
sampling = 1000

t = np.arange(0, duration, 1/sampling)
cosine = amp * np.cos(2 * np.pi * freq * t)
plt.plot(t, cosine)
plt.title("Cosine Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

t = np.linspace(-np.pi, np.pi, 500)
tangent = np.tan(t)
plt.plot(t, tangent)
plt.title("Tangent Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.ylim(-10, 10)
plt.grid(True)
plt.show()

t = np.linspace(-1, 1, 500)
impulse = np.zeros_like(t)
impulse[250] = 1
plt.stem(t, impulse, use_line_collection=True)
plt.title("Impulse Function")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

t = np.linspace(-1, 1, 500)
ramp = np.maximum(0, t)
plt.plot(t, ramp)
plt.title("Ramp Function")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

t = np.linspace(-1, 1, 500)
unit_step = np.heaviside(t, 1)
plt.plot(t, unit_step)
plt.title("Unit Step Function")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
